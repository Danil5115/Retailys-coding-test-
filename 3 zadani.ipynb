<p>3. Náhradní díly k jednotlivým produktům (jsou pouze u některých produktů).</p>
<p>Kód projde každý produkt, vyhledá sekci (parts), poté najde díly určité kategorie (v tomto případě kategorie "Náhradní díly" s identifikátorem 1) a zobrazí informace o produktu a jeho dílech v konzoli. U každého výrobku zkontroluje, zda existuje sekce parts, pak v této sekci vyhledá díly s požadovanou kategorií a vypíše jejich názvy a kódy</p>

```python
import xml.etree.ElementTree as ET

# Načtení souboru XML
tree = ET.parse('export_full.xml')
root = tree.getroot()

# Projdu jednotlivé produkty
for item in root.findall('.//item'):
    parts = item.find('.//parts')
    if parts is not None:
        part_category = parts.find('.//part[@categoryId="1"]')  # Získat údaje o kategorii dílů
        if part_category is not None:
            print(f'Product: {item.attrib["code"]} - {item.attrib["name"]}')
            # Projdu si detaily a zapíšu si jejich informace.
            for part in part_category.findall('.//item'):
                print(f'   Náhradní díl: {part.attrib["code"]} - {part.attrib["name"]}')
```
