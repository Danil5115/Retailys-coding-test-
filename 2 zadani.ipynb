<p>2. Názvy produktů vypíše jeden za druhým pod sebou.</p>
<p>Tento kód vyhledá všechny jedinečné názvy produktů v tomto souboru pomocí knihovny ElementTree. Při procházení každého "prvku" v souboru kód ukládá jejich názvy do množiny, čímž automaticky eliminuje opakující se názvy. Nakonec kód zobrazí všechny jedinečné názvy výrobků na obrazovce.</p>
<p>Kód je zde podobný, jako u předchozího úkolu, protože se v xml souboru opakují produkty v různých sekcích, hledám pouze jedinečné hodnoty a na konci vypisuji jejich názvy jeden po druhém.</p>

```python
import xml.etree.ElementTree as ET

# Parsování dokumentu XML
tree = ET.parse('export_full.xml')
root = tree.getroot()

# Vytvoření slovníku pro ukládání jedinečných produktů
unique_product_names = set()

# Prohledat všechny značky 'item' a uložit jedinečné názvy produktů
for item in root.findall('.//item'):
    name = item.get('name')
    unique_product_names.add(name)

# Zobrazení jedinečných názvů produktů
for name in unique_product_names:
    print(name)
```
